```python
# How many unique companies are in the dataset?

laptopData['Company'].nunique()

```

```python
# What are the most common types of laptops (TypeName)?

laptopData['TypeName'].value_counts().idxmax()
```

```python
# What is the range of prices in the dataset?
# range is the differences between max and min

# laptopData['Price'].describe()

max_range = laptopData['Price'].max()
# print(max_range, '\n')
min_range = laptopData['Price'].min()
# print(min_range, '\n')

Range = (max_range - min_range)
print(f'The range is {Range}')
```

```python
# What is the average price of laptops by brand?
# average is also referred to as mean
# laptopData['Price'].describe()

laptopData['Price'].mean()


```

```python
# Average price for each company prices

laptopData.groupby('Company')['Price'].mean()
```

```python
# How does the price of laptops vary by TypeName (e.g., Notebook, Ultrabook, etc.)?

laptopData.groupby('TypeName')['Price'].mean()
```
